```python
# prompt: read text from text file as a single string, removing new line characters
def read_text_file(file_path):
    with open(file_path, 'r') as file:
        file_contents = file.read().replace('\n', '')
    return file_contents

# Example usage
file_path = 'ex_8_3_corpus.txt'  # Replace with the path to your file
text = read_text_file(file_path)

print(text)
```

```python
import re

text = re.sub(r'[^\w\s]', '', text)
print(text)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
print(tokens)
```

```python
from collections import Counter

vocab = Counter(tokens).most_common(50)
print(vocab)
```

```python
word_to_index = {}
for i, (word, count) in enumerate(vocab):
    word_to_index[word] = i

tokens_index = []
for token in tokens:
  if token in word_to_index:
    tokens_index.append(word_to_index[token])
  else:
    tokens_index.append(-1)

print(tokens_index)
```

```python
# term-term co-occurrence matrix
import numpy as np

window_size = 4
num_words = len(vocab)
co_occurrence_matrix = np.zeros((num_words, num_words))

for i in range(len(tokens_index)):
    if tokens_index[i] == -1:
        continue
    for j in range(i - window_size, i + window_size + 1):
        if j < 0 or j >= len(tokens_index) or i == j or tokens_index[j] == -1:
            continue
        co_occurrence_matrix[tokens_index[i]][tokens_index[j]] += 1

print(co_occurrence_matrix)
```
