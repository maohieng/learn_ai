# Mini Project 3 - II. Dimensional Reduction
Apply dimensionality reduction using PCA and select only the two most important components from the word embedding, and plot each word representation on a 2D graph.

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

### Load Word Embeddings

```python
file_embeddings = "I_embeddings2.npy" # Change as you prefer
file_word_embbedings = "I_word_to_embeddings2.npy" # Change as you prefer
```

```python
embeddings2 = np.load(file_embeddings)
embeddings2
```

```python
word_to_embeddings2 = np.load(file_word_embbedings, allow_pickle=True).item()
word_to_embeddings2
```

### Apply PCA

```python
# Fit PCA on the embeddings
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings2)
```

```python
# Map the reduced embeddings to the words
reduced_word_embeddings = {word: reduced_embeddings[i] for i, word in enumerate(word_to_embeddings2.keys())}
reduced_word_embeddings
```

#### Plot the graph

```python
import pathlib
# from matplotlib.font_manager import FontProperties

# font = FontProperties(fname='KhmerOSContent-Regular.ttf')

# Plot the reduced embeddings
plt.figure(figsize=(20, 10))
for word, coord in reduced_word_embeddings.items():
    plt.scatter(coord[0], coord[1])
    plt.text(coord[0] + 0.01, coord[1] + 0.01, word, fontsize=9, font=pathlib.Path('KhmerOSContent-Regular.ttf'))
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('2D PCA of Word Embeddings')
plt.show()
```
