```python
import os
```

```python
# Read text file if exists instead of listdir
def read_file(file):
    try:
        with open(file) as f:
            return f.read().splitlines()
    except FileNotFoundError:
        return []
```

```python
img_path = '500images'
imgs = read_file('500images.txt')
if not imgs:
    print('No file found, listing directory')
    imgs = os.listdir(img_path)
imgs
```

```python
size = len(imgs)
size
```

```python
# Write the list of images to a file
with open('500images.txt', 'w') as f:
    for img in imgs:
        f.write(f'{img_path}/{img}\n')
```
