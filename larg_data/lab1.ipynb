## Initialize a SparkSession

```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Lab1") \
    .getOrCreate()
```

```python
print("Spark session initialized.", spark)
```

## Create a DataFrame

```python
# Sample data: list of tuples (name, department, salary)
data = [("Alice", "HR", 50000)]

# Define column names
columns = ["Name", "Department", "Salary"]

# Create a DataFrame
df = spark.createDataFrame(data, columns)
```

```python
df.printSchema()
```

```python
df.show()
```

## Perform Data Transformation

```python
# Filter employees with salary greater than 55000
high_earners = df.filter(df.Salary > 55000)
print("Employees with salary > 55000:")
high_earners.show()
```
