```python
x = [90, 130, 210, 300, 350, 420, 480, 530, 640, 710]
y = [7.1, 10.9, 19.2, 28, 32.8, 39.9, 46.1, 51, 62.2, 68.9]

print(len(x))
```

```python
# Plot x and y
import matplotlib.pyplot as plt

# Mark x-axis
plt.xlabel('x')
# Mark y-axis
plt.ylabel('y')
# Mark title
plt.title('Scatter Plot')

plt.scatter(x, y)
plt.show()
```

```python
# Hypothesis: h(x) = a*x
def h(x, a):
  return [a*xi for xi in x]

def h2(x, a, b):
  return [(a*xi + b) for xi in x]
```

```python
# Squared Error Function
def J(z, y):
  return sum([(zi - yi)**2 for zi, yi in zip(z, y)])/len(z)
```

```python
# GRADIENT DESCENT
def grad(x, y, a):
  return 2*sum([(a*xi - yi)*xi for xi, yi in zip(x, y)])/len(x)

def grada(x, y, a, b):
  return 2*sum([(a*xi + b - yi)*xi for xi, yi in zip(x, y)])/len(x)

def gradb(x, y, a, b):
  return 2*sum([(a*xi + b - yi) for xi, yi in zip(x, y)])/len(x)
```

## Train h(x) = a*x

```python
# Training h = a*x

a = 1000
alpha = 0.000001 # Learning rate
iter = 50

store_a = []
store_cost = []

for i in range(iter):
  g = grad(x, y, a)

  a += -alpha*g

  cost = J(h(x, a), y)
  store_a.append(a)
  store_cost.append(cost)
  print(f'{i+1}/{iter}: a = {a}, cost = {cost}')
```

```python
# Plot a and J
plt.xlabel('a')
plt.ylabel('J')
plt.title('Gradient Descent')
plt.plot(store_a, store_cost)
plt.show()
```

## Train h(x) = a*x + b

```python
# Miminum 609/1000: a = -1.9482111050389908, b = 983.6659243334894, cost = 202926.92830864116
# a = -1.92
# b = 985
# alpha = 0.0000053 # Learning rate

# Miminum 874/1000: a = -1.8912442608961015, b = 982.0880594718344, cost = 202261.51379812014
# a = -1.92
# b = 984
# alpha = 0.00000532 # Learning rate

# Miminum 120/1000: a = -1.8904886234732496, b = 981.7378155097249, cost = 202118.02478326688
a = -1.90
b = 982
alpha = 0.00000532 # Learning rate

# Final minimu cost value after experiments!!!
# Miminum 11/1000: a = -1.9410530905393204, b = 980.9758815002272, cost = 201804.93421222598
a = -1.89
b = 981
alpha = 0.00000532 # Learning rate

iter = 1000

store_a = []
store_b = []
store_cost = []

for i in range(iter):
  ga = grada(x, y, a, b)
  gb = gradb(x, y, a, b)

  a += -alpha*ga
  b += -alpha*gb

  cost = J(h2(x, a, b), y)
  store_a.append(a)
  store_cost.append(cost)
  store_b.append(b)
  print(f'{i+1}/{iter}: a = {a}, b = {b}, cost = {cost}')

min_cost = store_cost[0]
index = 0
for idx, val in enumerate(store_cost):
  if val < min_cost:
    min_cost = val
    index = idx

print(f'Miminum {index+1}/{iter}: a = {store_a[index]}, b = {store_b[index]}, cost = {min_cost}')
```
