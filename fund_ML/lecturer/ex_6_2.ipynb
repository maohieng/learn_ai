```python
import cv2

img = cv2.imread('/content/flower.jpg', cv2.IMREAD_COLOR)

print(img.shape)
```

```python
from sklearn.cluster import KMeans

x = img.copy()
x = x.reshape((-1, 3))
num_color = 3

model = KMeans(n_clusters=num_color)
model.fit(x)
```

```python
print('New Colors:\n', model.cluster_centers_)
print('Labels:\n', model.labels_[:10])
```

```python
for i in range(len(x)):
  x[i] = model.cluster_centers_[model.labels_[i]]

x = x.reshape(img.shape)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2)
axes[0].imshow(img)
axes[0].axis('off')
axes[1].imshow(x)
axes[1].axis('off')
plt.show()
```
