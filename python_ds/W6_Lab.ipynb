# Hand-On Activities 1: Plot Some Graphs

```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

## 1. Line Graph
Data: Monthly Sales Performance of a Product
- Month: January, February, March, April, May
- Sales: 1000, 1500, 1200, 1800, 2000

```python
month_sales = pd.DataFrame({'Months': ['January', 'Februry', 'March', 'April', 'May'], 
                            'Sales': [1000, 1500, 1200, 1800, 2000]})

plt.plot(month_sales['Months'], month_sales['Sales'], marker='o')
plt.title("Sales Performance")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

## 2. Bar Chart
Data: Annual Revenue Comparison of Multiple Products
- Product: A, B, C, D
- Revenue (in thousands): 250, 300, 400, 350

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
revenues = [250000, 300000, 400000, 350000]

plt.bar(products, revenues, color=['red', 'blue', 'green', 'orange'])
plt.title("Annual Revenue")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.show()
```

## 3. Scatter Plot
Data: Relationship between Study Hours and Test Scores
- Study Hours: 2, 4, 3, 6, 5
- Test Scores: 60, 70, 65, 80, 75

```python
study_hour_score = pd.DataFrame({'StudyHour': [2,4,3,6,5], 'Score': [60,70,65,80,75]})

plt.scatter(study_hour_score['StudyHour'], study_hour_score['Score'])
plt.title("Study Hour vs Score")
plt.xlabel("Study Hour")
plt.ylabel("Score")
plt.show()
```

## 4. Pie Chart
Data: Market Share of Different Mobile Brands
- Brand: Apple, Samsung, Xiaomi, Huawei, Others
- Market Share (in percentage): 40, 30, 15, 10, 5

```python
brand_markets = pd.DataFrame({'Brands': ['Apple', 'Samsung', 'Xiaomi', 'Huawei', 'Others'], 
                              'Share': [40, 30, 15, 10, 5]})

plt.pie(brand_markets['Share'], labels=brand_markets['Brands'], autopct='%1.1f%%')
plt.title("Smartphone Market Share")
plt.show()
```

## 5. Area Chart
Data: Daily Website Traffic Over a Week
- Day: Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday
- Traffic (in thousands): 2, 3, 4, 3.5, 4.5, 5, 2.5

```python
wekly_traffic = pd.DataFrame({'days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], 
                              'traffics': [2, 3, 4, 3.5, 4.5, 5, 2.5]})

plt.fill_between(wekly_traffic['days'], wekly_traffic['traffics'], color='skyblue', alpha=0.4)
plt.plot(wekly_traffic['days'], wekly_traffic['traffics'], color='Slateblue', alpha=0.6)
plt.title("Weekly Traffic")
plt.xlabel("Days")
plt.ylabel("Traffic (in thousands)")
plt.show()
```

## 6. Histogram
Data: Distribution of Students' Exam Scores
- Score Range: 60-69, 70-79, 80-89, 90-99
- Number of Students: 5, 10, 15, 8

```python
student_scores = pd.DataFrame({'Score_Range': ['60-69', '70-79', '80-89', '90-99'],
                                 'Number_of_Students': [5, 10, 15, 8]})

ax = student_scores.plot(kind='bar', x='Score_Range', y='Number_of_Students', color='skyblue')
ax.set_xlabel("Score Range")
ax.set_ylabel("Number of Students")
plt.title("Distribution of Students' Exam Scores")
plt.show()
```

## 7. Bubble Chart
Data: Population Comparison of Countries
- Country: USA, China, India, Brazil, Russia
- Population (in millions): 330, 1400, 1350, 210, 145

```python
cp_df = pd.DataFrame({'Country': ['USA', 'China', 'India', 'Brazil', 'Russia'],
                      'Population': [330, 1400, 1350, 210, 145]})

# Plot bubble chart
cp_df.plot(kind='scatter', x='Country', y='Population', s=cp_df['Population']*0.5, color='skyblue')
plt.title("Population Comparison of Countries")
plt.xlabel("Country")
plt.ylabel("Population (in million)")
plt.show()
```

## 8. Box and Whisker Plot
Data: Monthly Temperature Range in a City
- Month: January, February, March, April, May
- Temperature Range (in degrees Celsius): 5-10, 8-15,
10-18, 12-20, 15-25

```python
temp_data = pd.DataFrame({'Month': ['January', 'February', 'March', 'April', 'May'],
                            'Temperature_Range': ['5-10', '8-15', '10-18', '12-20', '15-25']})

# Extract min and max temperature
temp_data['Min_Temp'] = temp_data['Temperature_Range'].apply(lambda x: int(x.split('-')[0]))
temp_data['Max_Temp'] = temp_data['Temperature_Range'].apply(lambda x: int(x.split('-')[1]))

# Plot box plot
temp_data[['Min_Temp', 'Max_Temp']].plot(kind='box')
plt.title("Monthly Temperature Range")
plt.ylabel("Temperature (in degrees Celsius)")
plt.show()
```
