# XOR Perceptron

```python
# Implement a linear combination of two perceptrons of XOR gate
import numpy as np

def NOT(x):
    return int(not x)

def AND(x1, x2):
    return x1 and x2

def OR(x1, x2):
    return x1 or x2

def XOR(x1, x2):
    # return AND(OR(x1, x2), NOT(AND(x1, x2)))
    return OR(AND(x1, NOT(x2)), AND(NOT(x1), x2))

print(XOR(0, 0))
print(XOR(0, 1))
print(XOR(1, 0))
print(XOR(1, 1))
```

```python
import numpy as np
 
 
def sigmoid (X, alpha=1):
  return 1/(1+np.exp(-alpha * X))

def cal(X1, X2, alpha=1):
    #not operation
    wnot=-1
    thresholdnot=-0.5
    Snot=X1*wnot
    X=Snot-thresholdnot
    ynot1=sigmoid(X, alpha)
    # print("ynot1", ynot1)
    
    Snot=X2*wnot
    X=Snot-thresholdnot
    ynot2=sigmoid(X, alpha)
    # print("ynot2", ynot2)
    
    
    # and opertaion
    wand=1
    thresholdand=1.5
    Sand=X1*wand+X2*wand
    X=Sand-thresholdand
    yand=sigmoid(X, alpha)
    # print("yand", yand)

    # or operation
    wor=1
    thresholdor=0.5
    Sor=ynot1*wor+ynot2*wor
    X=Sor-thresholdor
    yor=sigmoid(X, alpha)   
    # print("yor", yor)

    # xor operation
    wxor=1
    thresholdxor=1.5
    Sxor=yand*wxor+yor*wxor
    X=Sxor-thresholdxor
    yxor=sigmoid(X, alpha)
    # print("yxor", yxor)

    return ynot1, ynot2, yand, yor, yxor

x1 = [0, 0, 1, 1]
x2 = [0, 1, 0, 1]
x1_not = []
x2_not = []
y_and = []
y_or = []
y_xor = []
for i in range(4):
    # print("X1", x1[i], "X2", x2[i])
    ynot1, ynot2, yand, yor, yxor = cal(x1[i], x2[i])
    x1_not.append(ynot1)
    x2_not.append(ynot2)
    y_and.append(yand)
    y_or.append(yor)
    y_xor.append(yxor)

import pandas as pd

df = pd.DataFrame({'X1': x1, 'X2': x2, 'X1_NOT': x1_not, 'X2_NOT': x2_not, 'AND': y_and, 'OR': y_or, 'XOR': y_xor})
df
```
